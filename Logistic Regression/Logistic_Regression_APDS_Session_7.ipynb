```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fname = "Pregnancy.csv"
train_data = pd.read_csv(fname)
```

```python
fname = "Pregnancy_Test.csv"
test_data = pd.read_csv(fname)
```

```python
train_data.drop(['Home/Apt/ PO Box_A'],axis=1,inplace=True)
```

```python
#train_data.drop(['Home/Apt/ PO Box_H'],axis=1,inplace=True)
train_data.drop(['Home/Apt/ PO Box_P'],axis=1,inplace=True)
```

```python
train_data.head()
```

```python
#One Hot Encoding for Gender and Home
train_data = pd.get_dummies(train_data,columns = ['Implied Gender'])
test_data = pd.get_dummies(test_data,columns = ['Implied Gender'])
```

```python
test_data.head()
```

```python
test_data.drop(['Home/Apt/ PO Box_P','Home/Apt/ PO Box_H','Home/Apt/ PO Box_P'],axis=1,inplace=True)
```

```python
test_data.drop(['Home/Apt/ PO Box_A'],axis=1,inplace=True)

test_data.head()
```

```python
test_data.head()
```

```python
test_data.info()
```

```python
train_data.columns
```

```python
#Create the train and test data
X_train = train_data.drop('PREGNANT',axis = 1)
y_train = train_data['PREGNANT']
```

```python
X_test = test_data.drop('PREGNANT',axis = 1)
y_test = test_data['PREGNANT']
```

```python
#import logistic regression
from sklearn.linear_model import LogisticRegression
```

```python
lr_model = LogisticRegression()
lr_model.fit(X_train,y_train)
```

```python
from sklearn.metrics import accuracy_score

predictions = lr_model.predict(X_test)
print(accuracy_score(predictions, y_test))
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, predictions)

print(cm)
```

```python
# visualize confusion matrix with seaborn heatmap

cm_matrix = pd.DataFrame(data=cm, columns=['Predicted -ve:0', 'Predicted +ve:1'], 
                                 index=['Actual -ve:0', 'Actual +ve:1'])

sns.heatmap(cm_matrix, annot=True, fmt='d', cmap='YlGnBu')
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test,predictions))
```

```python
 prediction_prob = lr_model.predict_proba(X_test)[:,1]

 prediction_prob[prediction_prob > 0.8] = 1
 prediction_prob[prediction_prob <= 0.8] = 0
 
```

```python
print(accuracy_score(y_test,prediction_prob))
```

```python
print(classification_report(y_test,prediction_prob))
```
